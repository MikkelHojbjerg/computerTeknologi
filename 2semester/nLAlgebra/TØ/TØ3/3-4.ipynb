```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
v = np.array([1.0, 1.0, 0.0, 1.0, 1.0]) [:, np.newaxis]
wt = np.array([[1.0, 1.0, 0.0, 1.0, 1.0, 1.0]]) 
v, wt
```

```python
fig, ax = plt.subplots()
ax.matshow(v @ wt, cmap = 'Reds', clim = (0.0, 1.2))
```
