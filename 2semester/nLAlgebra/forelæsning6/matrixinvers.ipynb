```python
import numpy as np
```

```python
a = np.array([[2.0, 3.0],
              [1.0, 2.0]])
a
```

```python
b = np.linalg.inv(a)
b
```

```python
b @ a
```

```python
a @ b
```

```python
c = np.array([[1.0, 1.0],
              [1.00000001, 1.0]])
c
```

```python
d @ c
```

```python
(d @ c) [0, 1]
```
