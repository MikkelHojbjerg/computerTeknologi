```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
A = [[2.0, -1.0], [-1.0, 2.0]]
```

```python
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.matshow(A)
```
