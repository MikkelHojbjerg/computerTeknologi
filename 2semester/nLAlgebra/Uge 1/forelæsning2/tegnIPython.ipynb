```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#punkterne x og y svarer til flere punkter (i dette tilf√¶lde 6 punkter) i et koordinatsystem 
x = np.array([-0.5, 0.0, 1.0, 2.0, 3.0, 4.0])
y = np.array([1.5, 2.0, 0.0, 1.5, 0.0, 2.0])
```

```python
x
```

```python
y
```

```python
fig, ax = plt.subplots()
#espect er med til at f√• orden i koordinatsystemtet
ax.set_aspect('equal')
ax.plot(x, y)
```

```python
#Graffen bliver nu flyttet
a = 1.0
b = 0.5
c = 0.0
d = 1.0
```

```python
fig, ax = plt.subplots()
ax.set_aspect('equal')
#Se forel√¶sning 2 slides (slide 13) for formel
ax.plot(a*x + b*y, c*x + d*y)
```

```python
#Graffen bliver roteret (se forl√¶sning 2, slide 15)
a = np.cos(np.pi/3)
b = -np.sin(np.pi/3)
c = np.sin(np.pi/3)
d = np.cos(np.pi/3)
```

```python
fig, ax = plt.subplots()
ax.set_aspect('equal')
#Se forel√¶sning 2 slides (slide 13) for formel
ax.plot(a*x + b*y, c*x + d*y)
```

```python
#Graffen gemmes
fig.savefig("fig-rotation.pdf")
```

```python
#Skift farve p√• firgur
fig, ax = plt.subplots()
ax.set_aspect('equal')
#Se forel√¶sning 2 slides (slide 13) for formel
ax.plot(a*x + b*y, c*x + d*y, color="red")
```

```python
#Marker punkter i koordinatsystemet
fig, ax = plt.subplots()
ax.set_aspect('equal')
#Se forel√¶sning 2 slides (slide 13) for formel
ax.plot(a*x + b*y, c*x + d*y, marker="o")
```

```python
#plot af funktioner (i dette tilf√¶lde cos)
t = np.linspace(0, 2*np.pi, 100)
t
```

```python
type(t)
```

```python
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.plot(t, np.cos(t))
```

```python
fig, ax = plt.subplots()
ax.set_aspect('equal')
#Plot cos i x retning og sin i y retning (dette giver enhedscirklen)
ax.plot(np.cos(t), np.sin(t))
```

```python
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-2.5, 2.5)
#Se forel√¶sning 2 slides (slide 13) for formel
ax.plot(np.cos(t), np.sin(t))
```
